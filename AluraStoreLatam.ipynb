### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda4.head()
```

#1. Análisis de facturación



```python

def calcular_ingresos(tienda: pd.DataFrame) -> float:
    return tienda['Precio'].sum()

ingresosTienda = calcular_ingresos(tienda)
ingresosTienda2 = calcular_ingresos(tienda2)
ingresosTienda3 = calcular_ingresos(tienda3)
ingresosTienda4 = calcular_ingresos(tienda4)

```

```python
print(f'ingresos de la Tienda 1 = {ingresosTienda}')
print(f'ingresos de la Tienda 2 = {ingresosTienda2}')
print(f'ingresos de la Tienda 3 = {ingresosTienda3}')
print(f'ingresos de la Tienda 4 = {ingresosTienda4}')

listaTiendaIngresos = [ingresosTienda, ingresosTienda2, ingresosTienda3, ingresosTienda4]
listaTiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(x= listaTiendas, height=listaTiendaIngresos)
plt.show()
```

# 2. Ventas por categoría

```python
def ventas_por_categoria_tienda(tienda_df, nombre_tienda):
    ventas_categoria = tienda_df['Categoría del Producto'].value_counts()

    print(f"\nVentas por categoría en {nombre_tienda}:\n")
    print(ventas_categoria)

    # Gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(ventas_categoria, labels=ventas_categoria.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribución de ventas por categoría en {nombre_tienda}')
    plt.axis('equal')
    plt.show()

# Aplicar la función a cada tienda
ventas_por_categoria_tienda(tienda, 'Tienda 1')
ventas_por_categoria_tienda(tienda2, 'Tienda 2')
ventas_por_categoria_tienda(tienda3, 'Tienda 3')
ventas_por_categoria_tienda(tienda4, 'Tienda 4')
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tienda: pd.DataFrame) -> float:
    return round(tienda['Calificación'].mean(), 4)
```

```python
calificacionTienda = calificacion_promedio(tienda)
calificacionTienda2 = calificacion_promedio(tienda2)
calificacionTienda3 = calificacion_promedio(tienda3)
calificacionTienda4 = calificacion_promedio(tienda4)

print(f'Calificación promedio de la Tienda 1 = {calificacionTienda}')
print(f'Calificación promedio de la Tienda 2 = {calificacionTienda2}')
print(f'Calificación promedio de la Tienda 3 = {calificacionTienda3}')
print(f'Calificación promedio de la Tienda 4 = {calificacionTienda4}')

listaTiendaCalificacion = [calificacionTienda, calificacionTienda2, calificacionTienda3, calificacionTienda4]
listaTiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.plot(listaTiendas, listaTiendaCalificacion, marker='o')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio del Cliente')
plt.title('Calificación Promedio del Cliente por Tienda')
plt.grid(True)
plt.show()
```

# 4. Productos más y menos vendidos

```python

def productos_mas_y_menos_vendidos(tienda_df: pd.DataFrame, nombre_tienda: str):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto').size().sort_values(ascending=False)

    print(f"\nProductos más vendidos en {nombre_tienda}:")
    print(ventas_por_producto.head())

    print(f"\nProductos menos vendidos en {nombre_tienda}:")
    print(ventas_por_producto.tail())

    # Visualización
    plt.figure(figsize=(12, 6))

    # Productos más vendidos
    ventas_por_producto.head().plot(kind='bar', color='skyblue', ax=plt.gca())
    plt.title(f'Top 5 Productos Más Vendidos en {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Número de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Productos menos vendidos
    plt.figure(figsize=(12, 6))
    ventas_por_producto.tail().plot(kind='bar', color='salmon', ax=plt.gca())
    plt.title(f'Top 5 Productos Menos Vendidos en {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Número de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

```

```python

# Aplicar la función a cada tienda
productos_mas_y_menos_vendidos(tienda, 'Tienda 1')
productos_mas_y_menos_vendidos(tienda2, 'Tienda 2')
productos_mas_y_menos_vendidos(tienda3, 'Tienda 3')
productos_mas_y_menos_vendidos(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
def costo_envio_promedio(tienda: pd.DataFrame) -> float:
    return round(tienda['Costo de envío'].mean(), 2)

costoEnvioTienda1 = costo_envio_promedio(tienda)
costoEnvioTienda2 = costo_envio_promedio(tienda2)
costoEnvioTienda3 = costo_envio_promedio(tienda3)
costoEnvioTienda4 = costo_envio_promedio(tienda4)

print(f'Costo de envío promedio de la Tienda 1 = {costoEnvioTienda1}')
print(f'Costo de envío promedio de la Tienda 2 = {costoEnvioTienda2}')
print(f'Costo de envío promedio de la Tienda 3 = {costoEnvioTienda3}')
print(f'Costo de envío promedio de la Tienda 4 = {costoEnvioTienda4}')

costos_envio = {
    'Tienda 1': costoEnvioTienda1,
    'Tienda 2': costoEnvioTienda2,
    'Tienda 3': costoEnvioTienda3,
    'Tienda 4': costoEnvioTienda4
}

tiendas = list(costos_envio.keys())
promedios = list(costos_envio.values())

plt.figure(figsize=(8, 6))
plt.bar(x = tiendas, height=promedios, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.ylim(0, max(promedios) * 1.2)
for i, v in enumerate(promedios):
    plt.text(i, v + 0.1, str(v), ha='center')
plt.show()
```
